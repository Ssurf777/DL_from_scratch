```python
import numpy as np

class Sigmoid:
    def forward(self, x, w, b):
        """
        x.shape = (データ数, 次元数)
        w.shape = (1, 次元数)
        b.shape = (1,)
        """
        self.x = x
        z = np.sum(w * x, axis=1) + b  # 次元方向にsum
        y_pred = 1 / (1 + np.exp(-z))  # sigmoid関数
        self.y_pred = y_pred
        return y_pred

    def backward(self, dy):
        """
        dy.shape = (データ数, )
        """
        dz = dy * (1.0 - self.y_pred) * self.y_pred # 問
        dw = np.sum(dz.T * self.x, axis=0) # データ数方向にsum
        db = np.sum(dz)
        return dw, db

class NegativeLogLikelihood:
     def forward(self, y_pred, y_true):
        """
        負の対数尤度
        y_pred.shape = (データ数, )
        y_ture.shape = (データ数, )
        """
        self.y_pred = y_pred
        self.y_true = y_true
        loss = -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) # 問
        return loss.sum()

    def backward(self):
        dy = -(self.y_true / self.y_pred) + ( (1 - self.y_true) / (1 - self.y_pred) ) # 問 d(loss)/d(y_pred)で微分
        return dy
```

# forward解説
* xは入力データ、wは重み、bはバイアスです。
* zは線形結合の結果で、z = w・x + bという形になります。
* y_predはシグモイド関数を通して得られる予測値です。
